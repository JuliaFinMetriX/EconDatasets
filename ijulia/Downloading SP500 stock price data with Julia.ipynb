```julia
using Taro
Taro.init()
using EconDatasets
using Datetime
```

```julia
url = "http://us.spindices.com/idsexport/file.xls?hostIdentifier=48190c8c-42c4-46af-8d1a-0cd5db894797&selectedModule=Constituents&selectedSubModule=ConstituentsFullList&indexId=340"
filepath = download(url)
constituents = Taro.readxl(filepath, "Constituents", "A10:B511")
```

```julia
(nTicker, nVars) = size(constituents)
```

```julia
ticker = [ii for ii in constituents[:Symbol]]
for ii=1:nTicker
    if isa(ticker[ii], Bool)
        println(ii)
    end
end
```

```julia
companyNames = ASCIIString[compName for compName in constituents[:Constituent]]
```

```julia
for ii=1:nTicker
    if ismatch(r"AT&T", companyNames[ii])
        # println(constituents[ii, 2])
        constituents[ii, 2] = "T"
    end
    
    if ismatch(r"Ford", companyNames[ii])
        # println(constituents[ii, 2])
        constituents[ii, 2] = "F"
    end
end
```

```julia
tickerSymb = ASCIIString[tickSymb for tickSymb in constituents[:Symbol]]
```

```julia
dates = date(1960,1,1):date(2014,7,20)
@time vals = EconDatasets.readYahooFinance(dates, tickerSymb, :d)
```

```julia
typeof(vals)
```

```julia
pathToStore = joinpath(Pkg.dir("EconDatasets"), "data", "SP500")
writeTimedata(pathToStore, vals)
```

```julia
(nObs, nStocks) = size(vals)

NAorNot = Array(Int, nObs, nStocks)

for ii=1:nStocks
    NAorNot[:, ii] = isna(vals.vals[ii])*1
end
```

```julia
NAorNot
```

```julia
nNAs = sum(NAorNot, 1)
p = sortperm(nNAs[:])
tickSorted = tickerSymb[p]
NAorNotSorted = NAorNot[:, p]
```

```julia
using Plotly

Plotly.signin("cgroll", "2it4121bd9")
nToPlot = 501
data = [
  [
        "z" => NAorNotSorted[:, 1:nToPlot]', 
        "x" => tickSorted[1:nToPlot], 
     "type" => "heatmap"
  ]
]

response = Plotly.plot([data], ["filename" => "labelled-heatmap", "fileopt" => "overwrite"])
plot_url = response["url"]
```
